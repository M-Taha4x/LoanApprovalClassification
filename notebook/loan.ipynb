__LOG TRANSFORM__

```python
df['person_income']=np.log1p(df['person_income'])
```

__ENCODING__

Label Encoding


```python
from sklearn.preprocessing import LabelEncoder
```

```python
le=LabelEncoder()
```

```python
df['person_gender']=le.fit_transform(df['person_gender'])
df['previous_loan_defaults_on_file']=le.fit_transform(df['previous_loan_defaults_on_file'])
```

Get Dummies

```python
df=pd.get_dummies(
    df,
    columns=['loan_intent'],
    dtype=int
)
```

```python
df=pd.get_dummies(
    df,
    columns=['person_home_ownership'],
    dtype=int
)
```

```python
df=pd.get_dummies(
    df,
    columns=['person_education'],
    dtype=int
)
```

__SPLITTING FEATURES__

```python
x=df.drop('loan_status',axis=1)
```

__SPLITTING TARGET VARIABLES__

```python
y=df['loan_status']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.25,random_state=42,stratify=y)
```

__REDEFINING NUMERICAL COLUMNS__

```python
num_col=x.select_dtypes(include=['int64','float64']).columns
```

__SCALING__

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
X_train[num_col]=scaler.fit_transform(X_train[num_col])
X_test[num_col]=scaler.transform(X_test[num_col])
```

____MODEL TRAINING____

__Logistic Regression__

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
```

```python
lr.fit(X_train,Y_train)
```

```python
pred_lr=lr.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,roc_auc_score
```

```python
acc_lr=accuracy_score(Y_test,pred_lr)
p_lr=precision_score(Y_test,pred_lr)
f1_lr=f1_score(Y_test,pred_lr)
rec_lr=recall_score(Y_test,pred_lr)
rc_lr=roc_auc_score(Y_test,pred_lr)
cm_lr=confusion_matrix(Y_test,pred_lr)
print("     Logistic Regression     ")
print("Accuracy Score: ",acc_lr)
print("Precision Score: ",p_lr)
print("F1 Score: ",f1_lr)
print("Recall Score: ",rec_lr)
print("Roc-Auc Score: ",rc_lr)
print("Confusion Matrix: \n",cm_lr)
```

__Decision Tree__

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt=DecisionTreeClassifier(criterion='gini',max_depth=None)
```

```python
dt.fit(X_train,Y_train)
```

```python
pred_dt=dt.predict(X_test)
```

```python
acc_dt=accuracy_score(Y_test,pred_dt)
p_dt=precision_score(Y_test,pred_dt)
f1_dt=f1_score(Y_test,pred_dt)
rec_dt=recall_score(Y_test,pred_dt)
rc_dt=roc_auc_score(Y_test,pred_dt)
cm_dt=confusion_matrix(Y_test,pred_dt)
print("     Decision Tree Classifier     ")
print("Accuracy Score: ",acc_dt)
print("Precision Score: ",p_dt)
print("F1 Score: ",f1_dt)
print("Recall Score: ",rec_dt)
print("Roc-Auc Score: ",rc_dt)
print("Confusion Matrix: \n",cm_dt)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf=RandomForestClassifier(n_estimators=100,random_state=42)
```

```python
rf.fit(X_train,Y_train)
```

```python
pred_rf=rf.predict(X_test)
```

```python
acc_rf=accuracy_score(Y_test,pred_rf)
p_rf=precision_score(Y_test,pred_rf)
f1_rf=f1_score(Y_test,pred_rf)
rec_rf=recall_score(Y_test,pred_rf)
rc_rf=roc_auc_score(Y_test,pred_rf)
cm_rf=confusion_matrix(Y_test,pred_rf)
print("     Random Forest Classifier     ")
print("Accuracy Score: ",acc_rf)
print("Precision Score: ",p_rf)
print("F1 Score: ",f1_rf)
print("Recall Score: ",rec_rf)
print("Roc-Auc Score: ",rc_rf)
print("Confusion Matrix: \n",cm_rf)
```

__KNN__

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(n_neighbors=5,metric='minkowski',p=2)
```

```python
knn.fit(X_train,Y_train)
```

```python
pred_knn=knn.predict(X_test)
```

```python
acc_knn=accuracy_score(Y_test,pred_knn)
p_knn=precision_score(Y_test,pred_knn)
f1_knn=f1_score(Y_test,pred_knn)
rec_knn=recall_score(Y_test,pred_knn)
rc_knn=roc_auc_score(Y_test,pred_knn)
cm_knn=confusion_matrix(Y_test,pred_knn)
print("     K-Nearest Neighbors     ")
print("Accuracy Score: ",acc_knn)
print("Precision Score: ",p_knn)
print("F1 Score: ",f1_knn)
print("Recall Score: ",rec_knn)
print("Roc-Auc Score: ",rc_knn)
print("Confusion Matrix: \n",cm_knn)
```

__XGBOOST__

```python
from xgboost import XGBClassifier
```

```python
xg=XGBClassifier(n_estimators=100,learning_rate=0.1,max_depth=5,objective='binary:logistic',random_state=42)
```

```python
xg.fit(X_train,Y_train)
```

```python
pred_xg=xg.predict(X_test)
```

```python
acc_xg=accuracy_score(Y_test,pred_xg)
p_xg=precision_score(Y_test,pred_xg)
f1_xg=f1_score(Y_test,pred_xg)
rec_xg=recall_score(Y_test,pred_xg)
rc_xg=roc_auc_score(Y_test,pred_xg)
cm_xg=confusion_matrix(Y_test,pred_xg)
print("     XGB Classifier    ")
print("Accuracy Score: ",acc_xg)
print("Precision Score: ",p_xg)
print("F1 Score: ",f1_xg)
print("Recall Score: ",rec_xg)
print("Roc-Auc Score: ",rc_xg)
print("Confusion Matrix: \n",cm_xg)
```

```python
result=pd.DataFrame({
    "Model":[
        "Logistic Regression",
        "Decision Tree Classifier",
        "Random Forest Classifier",
        "K-Nearest Neighbors",
        "XGB Classifier"
    ],
    "Accuracy":[
        acc_lr,acc_dt,acc_rf,acc_knn,acc_xg
    ],
    "Precision":[
        p_lr,p_dt,p_rf,p_knn,p_xg
    ],
    "Recall":[
        rec_lr,rec_dt,rec_rf,rec_knn,rec_xg
    ],
    "F1 Scorre":[
        f1_lr,f1_dt,f1_rf,f1_knn,f1_xg
    ],
    "ROC-AUC Score":[
        rc_lr,rc_dt,rc_rf,rc_knn,rc_xg
    ]
})
    
```
